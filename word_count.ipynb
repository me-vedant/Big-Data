```python
import findspark
findspark.init('/usr/local/spark')
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
from pyspark.sql import Row
```

```python
sc =  spark.sparkContext
```

```python
c = sc.textFile("/home/hduser/Downloads/sharedfolder/war_and_peace.txt")
```

```python
c.collect()
```

```python
cfmap = c.flatMap(lambda x : x.split(" "))
```

```python
cmap = cfmap.map(lambda x: (x,1))
```

```python
count = cmap.reduceByKey(lambda x,y : (x+y))
```

```python
count.collect()
```
